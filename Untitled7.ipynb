```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x))
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),label='cos')
ax.set_title('multiple lines')
ax.set_xlabel('periods')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()

```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x),label='sin',color='green')
ax.plot(x,np.cos(x),label='cos',color='black')
ax.set_title('multiple lines')
ax.set_xlabel('periods')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='sin x+1',linestyle='dashdot')
ax.set_title('multiple lines')
ax.set_xlabel('periods')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()

```

```python
fig,axs = plt.subplots(2, 2, figsize=(10,6))
x = np.linspace(0, 10, 1000)
axs[0,0].plot(x, np.sin(x))
axs[0,1].plot(x, np.cos(x), color='maroon')
axs[1,0].plot(x, np.sin(x+2), color='blue')
axs[1,1].plot(x, np.sin(x+4), color='green')
```

```python
data={'Apple':10,'Orange':15,'Lemon':5,'Banana':13,'Pear':9,'Pomegranate':11,'Lime':2}
names=list(data.keys())
values=list(data.values())
values.sort()
ax=plt.axes()
ax.barh(names,values);
```

```python
data={'Apple':10,'Orange':15,'Lemon':5,'Banana':13,'Pear':9,'Pomegranate':11,'Lime':2}
names=list(data.keys())
values=list(data.values())
ax=plt.axes()
ax.barh(names,values)
ax.set_xlabel('Quantity')
ax.set_ylabel('Fruit')
ax.set_title('OMG')
```

```python
labels=['G1','G2','G3','G4','G5']
men_means=[20,34,30,35,27]
women_means=[25,32,34,20,25]
x=np.arange(len(labels))
width=0.25
fig,ax=plt.subplots()
rects1=ax.bar(x-width/2,men_means,width,label='Men')
rects2=ax.bar(x+width/2,women_means,width,label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

```python
labels=['G1','G2','G3','G4','G5']
men_means=[20,34,30,35,27]
women_means=[25,32,34,20,25]
x=np.arange(len(labels))
width=0.25
fig,ax=plt.subplots()
rects1=ax.bar(x-width/2,men_means,width,label='Men')
rects2=ax.bar(x+width/2,women_means,width,label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.bar_label(rects1,padding=3)
ax.bar_label(rects2,padding=3)
ax.set_ylim(0,50);


```
